```python
import pandas as pd

df = pd.read_csv('finalised/BNF_5LS_final.csv')
df
```

```python
import pandas as pd
df['Date'] = pd.to_datetime(df['Date'])
df['PnL_w_cs'] = df['PnL'] - (0.001 * df['Quantity'] * -1 * (df['Entry Price'] + df['Exit Price']))
df['ROI_w_cs'] = df['PnL_w_cs'] / 10000
daily_pnl = df.groupby(df['Date'].dt.date)['PnL_w_cs'].sum().reset_index()

# Rename the columns for clarity
daily_pnl.columns = ['Date', 'Total PnL']

daily_pnl["Date"] = pd.to_datetime(daily_pnl["Date"])
daily_pnl['Trade Year'] = daily_pnl['Date'].dt.year
daily_pnl["ROI"] = daily_pnl["Total PnL"]/10000
# If you want to ensure 'Date' column is in datetime format
daily_pnl['Date'] = pd.to_datetime(daily_pnl['Date'])

# Display the resulting DataFrame
print(daily_pnl)
```

```python
df["Date"] = pd.to_datetime(df["Date"])
print(type(df["Date"]))
df['Trade Year'] = df['Date'].dt.year
df
```

```python
df = daily_pnl
stats_ = pd.DataFrame(index=range(2017, 2024), columns=['Total ROI', 'Total Trades', 'Win Rate', 'Avg Profit% per Trade', 'Avg Loss% per Trade', 'Max Drawdown', 'ROI/DD Ratio'])

# Iterate over each year
for year in range(2017, 2024):
    # Filter trades for the current year
    year_trades = df[(df['Trade Year'] == year)]
    
    # Calculate total ROI
    total_roi = year_trades['ROI_w_cs'].sum()
    
    # Calculate total number of trades
    total_trades = len(year_trades)
    
    # Calculate win rate
    win_rate = (year_trades['ROI_w_cs'] > 0).mean()*100
    
    # Calculate average profit per trade
    avg_profit = year_trades[year_trades['ROI'] > 0]['ROI_w_cs'].mean()
    
    # Calculate average loss per trade
    avg_loss = year_trades[year_trades['ROI_w_cs'] < 0]['ROI'].mean()
    
    # Calculate maximum drawdown
    max_drawdown = (year_trades['ROI_w_cs'].cumsum() - year_trades['ROI_w_cs'].cumsum().cummax()).min()
    
    # Calculate ROI/DD ratio
    roi_dd_ratio = total_roi / abs(max_drawdown)
    # 
    # Store the statistics in the DataFrame
    stats_.loc[year] = [total_roi, total_trades, win_rate, avg_profit, avg_loss, max_drawdown, roi_dd_ratio]

# Calculate overall statistics
overall_total_roi = stats_['Total ROI'].sum()
overall_total_trades = stats_['Total Trades'].sum()
overall_win_rate = (df['ROI_w_cs'] > 0).mean() * 100
overall_avg_profit = df[df['ROI_w_cs'] > 0]['ROI_w_cs'].mean()
overall_avg_loss = df[df['ROI_w_cs'] < 0]['ROI_w_cs'].mean()
overall_max_drawdown = (df['ROI_w_cs'].cumsum() - df['ROI_w_cs'].cumsum().cummax()).min()
overall_roi_dd_ratio = overall_total_roi / abs(overall_max_drawdown)

# Store the overall statistics in the DataFrame
stats_.loc['Overall'] = [overall_total_roi, overall_total_trades, overall_win_rate, overall_avg_profit, overall_avg_loss, overall_max_drawdown, overall_roi_dd_ratio]
stats_
```
