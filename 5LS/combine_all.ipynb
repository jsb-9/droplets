```python
import pandas as pd

df1 = pd.read_csv("finalised/FNF_5LS_final.csv")
df2 = pd.read_csv("finalised/BNF_5LS_final.csv")
df3 = pd.read_csv("finalised/NF_5LS_final.csv")
```

```python
list_bnf = []
list_nf = []

for i in range(1, len(df2)):
    expiry = df2.iloc[i]['Expiry']
    list_bnf.append(expiry)

for i in range(1, len(df3)):
    expiry = df3.iloc[i]['Expiry']
    list_nf.append(expiry)

list_bnf = list(set(list_bnf))
list_nf = list(set(list_nf))
```

```python
list_bnf
```

```python
list_nf
```

```python
for i in range(1, len(df2)):
    if df2.iloc[i]['Expiry'] in list_nf:
        df2.iloc[i]['Quantity'] = df2.iloc[i]['Quantity'] / 2
        df2.iloc[i]['PnL'] = df2.iloc[i]['PnL'] / 2
        df2.iloc[i]['ROI'] = df2.iloc[i]['ROI'] / 2

for i in range(1, len(df3)):
    if df2.iloc[i]['Expiry'] in list_bnf:
        df2.iloc[i]['Quantity'] = df2.iloc[i]['Quantity'] / 2
        df2.iloc[i]['PnL'] = df2.iloc[i]['PnL'] / 2
        df2.iloc[i]['ROI'] = df2.iloc[i]['ROI'] / 2
```

```python
final_df = pd.concat([df1, df2, df3])
final_df
```

```python
final_df['Datetime'] = pd.to_datetime(final_df['Date'] + ' ' + final_df['Entry Time'])
final_df
```

```python
final_df_sorted = final_df.sort_values(by='Datetime')
final_df_sorted.reset_index(drop=True, inplace=True)
final_df_sorted
```

```python
final_df_sorted.to_csv('All Indices 5LS V_1.1.csv')
```
