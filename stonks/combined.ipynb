```python
def stats(stock_name, df):
    print(stock_name)
    stats = pd.DataFrame(index=range(2011, 2025), columns=['Total ROI', 'Total Trades', 'Win Rate', 'Avg Profit% per Trade', 'Avg Loss% per Trade', 'Max Drawdown', 'ROI/DD Ratio'])
    # df['ROI%'] = df['ROI%']/5
    # Iterate over each year
    for year in range(2011, 2025):
        # Filter trades for the current year
        year_trades = df[(df['Trade Year'] == year)]
        
        # Calculate total ROI
        total_roi = year_trades['Combined ROI'].sum()
        
        # Calculate total number of trades
        total_trades = len(year_trades)
        
        # Calculate win rate
        win_rate = (year_trades['Combined ROI'] > 0).mean()*100
        
        # Calculate average profit per trade
        avg_profit = year_trades[year_trades['Combined ROI'] > 0]['Combined ROI'].mean()
        
        # Calculate average loss per trade
        avg_loss = year_trades[year_trades['Combined ROI'] < 0]['Combined ROI'].mean()
        
        # Calculate maximum drawdown
        max_drawdown = (year_trades['Combined ROI'].cumsum() - year_trades['Combined ROI'].cumsum().cummax()).min()
        
        # Calculate ROI/DD ratio
        roi_dd_ratio = total_roi / abs(max_drawdown)
        
        # Store the statistics in the DataFrame
        stats.loc[year] = [total_roi, total_trades, win_rate, avg_profit, avg_loss, max_drawdown, roi_dd_ratio]
    
    # Calculate overall statistics
    overall_total_roi = stats['Total ROI'].sum()
    overall_total_trades = stats['Total Trades'].sum()
    overall_win_rate = (df['Combined ROI'] > 0).mean() * 100
    overall_avg_profit = df[df['Combined ROI'] > 0]['Combined ROI'].mean()
    overall_avg_loss = df[df['Combined ROI'] < 0]['Combined ROI'].mean()
    overall_max_drawdown = (df['Combined ROI'].cumsum() - df['Combined ROI'].cumsum().cummax()).min()
    overall_roi_dd_ratio = overall_total_roi / abs(overall_max_drawdown)
    
    # Store the overall statistics in the DataFrame
    stats.loc['Overall'] = [overall_total_roi, overall_total_trades, overall_win_rate, overall_avg_profit, overall_avg_loss, overall_max_drawdown, overall_roi_dd_ratio]
    return stats

results = stats('nifty nxt50 16', combined_df)
results
```

```python
import pandas as pd

def stats(stock_name, df):
    print(stock_name)
    stats = pd.DataFrame(index=pd.MultiIndex.from_product([range(2011, 2025), range(1, 13)], names=['Year', 'Month']),
                         columns=['Total ROI', 'Total Trades', 'Win Rate', 'Avg Profit% per Trade', 'Avg Loss% per Trade', 'Max Drawdown', 'ROI/DD Ratio'])
    
    # Iterate over each year and month
    for year in range(2011, 2025):
        for month in range(1, 13):
            # Filter trades for the current year and month
            year_month_trades = df[(df['Trade Year'] == year) & (df['Trade Month'] == month)]
            
            # Calculate total ROI
            total_roi = year_month_trades['Combined ROI'].sum()
            
            # Calculate total number of trades
            total_trades = len(year_month_trades)
            
            if total_trades > 0:
                # Calculate win rate
                win_rate = (year_month_trades['Combined ROI'] > 0).mean() * 100
                
                # Calculate average profit per trade
                avg_profit = year_month_trades[year_month_trades['Combined ROI'] > 0]['Combined ROI'].mean()
                
                # Calculate average loss per trade
                avg_loss = year_month_trades[year_month_trades['Combined ROI'] < 0]['Combined ROI'].mean()
                
                # Calculate maximum drawdown
                max_drawdown = (year_month_trades['Combined ROI'].cumsum() - year_month_trades['Combined ROI'].cumsum().cummax()).min()
                
                # Calculate ROI/DD ratio
                roi_dd_ratio = total_roi / abs(max_drawdown)
            else:
                win_rate = 0
                avg_profit = 0
                avg_loss = 0
                max_drawdown = 0
                roi_dd_ratio = 0
            
            # Store the statistics in the DataFrame
            stats.loc[(year, month)] = [total_roi, total_trades, win_rate, avg_profit, avg_loss, max_drawdown, roi_dd_ratio]
    
    # Calculate overall statistics
    overall_total_roi = stats['Total ROI'].sum()
    overall_total_trades = stats['Total Trades'].sum()
    overall_win_rate = (df['Combined ROI'] > 0).mean() * 100
    overall_avg_profit = df[df['Combined ROI'] > 0]['Combined ROI'].mean()
    overall_avg_loss = df[df['Combined ROI'] < 0]['Combined ROI'].mean()
    overall_max_drawdown = (df['Combined ROI'].cumsum() - df['Combined ROI'].cumsum().cummax()).min()
    overall_roi_dd_ratio = overall_total_roi / abs(overall_max_drawdown)
    
    # Store the overall statistics in the DataFrame
    stats.loc['Overall'] = [overall_total_roi, overall_total_trades, overall_win_rate, overall_avg_profit, overall_avg_loss, overall_max_drawdown, overall_roi_dd_ratio]
    
    return stats

# Call the function for every month of every year
results = stats('nifty nxt50 16', combined_df)
results
```
