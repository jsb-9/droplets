```python
from datetime import timedelta

def trade_next_open_monthly_long(stock_name, df):
    in_trade = False
    entry_price = 0
    exit_price = 0
    signal_candle_high = 0
    points = 0
    remark = ''
    initial_sl_pct = 100
    target_pct = 1500
    portfolio_value = 10_00_000
    trade_number = 1

    tradebook = []
    
    for i in range(0, len(df)):
        
        if (df.iloc[i]['monthly_long'] == 1) and not in_trade:
            #Entry Triggered
            in_trade = True
            entry_price = df.iloc[i]['close']
            entry_time = df.iloc[i]['datetime']
            initial_sl = entry_price * (1 - (initial_sl_pct / 100))
            initial_target = entry_price * (1 + (target_pct / 100))
    
            # print('Entry Triggered ! ', df.iloc[i]['datetime'])

            continue

            #TODO : Try using the nearer SL between low of entry candle and initial_sl fixed for Initial SL of the trade
            # ----> initial_sl = max( df.iloc[i]['low'] , entry_price * (1 - (initial_sl_pct / 100)) )
            
        if in_trade:
            if df.iloc[i]['monthly_long'] == -1:
                # Supertrend reversed, SL hit
                exit_price = df.iloc[i]['close']
                exit_time = df.iloc[i]['datetime']
                in_trade = False
                points = exit_price - entry_price
                remark = 'SuperTrend Reversed'
                # print(remark, df.iloc[i]['datetime'])

            elif df.iloc[i]['low'] < initial_sl:
                #Initial SL hit
                
                if df.iloc[i]['open'] < initial_sl:
                    #Gapped below Initial SL, exit at open
                    exit_price = df.iloc[i]['open']
                    exit_time = df.iloc[i]['datetime']
                    remark = 'Gapped Outside Initial SL'
                    # print(remark, df.iloc[i]['datetime'])

                else:
                    exit_price = initial_sl
                    exit_time = df.iloc[i]['datetime']
                    remark = 'Initial SL Hit'
                    print(remark, df.iloc[i]['datetime'])
                    
                in_trade = False
                points = exit_price - entry_price

            elif df.iloc[i]['high'] > initial_target:
                # Target Hit
                exit_price = initial_target
                exit_time = df.iloc[i]['datetime']
                in_trade = False
                points = exit_price - entry_price
                remark = 'Target Hit'
                # print(remark, df.iloc[i]['datetime'])

            elif df.iloc[i]['indexer'] == df.iloc[-1]['indexer']:
                # Ending Currently Active Trade
                exit_price = df.iloc[i]['close']
                exit_time = df.iloc[i]['datetime']
                in_trade = False
                points = exit_price - entry_price
                remark = 'Currently Active Trade'
                # print(remark, df.iloc[i]['datetime'])

        if points:
            # Calculate PnL for the whole trade duration
            qty = portfolio_value / entry_price
            slippage = 0.001 * (entry_price + exit_price)
            # final_points = points - slippage
            
            trade_duration = (exit_time - entry_time)

            st_index = df.loc[df['datetime'] == entry_time, 'indexer'].iloc[0]
            en_index = df.loc[df['datetime'] == exit_time, 'indexer'].iloc[0]
            current_index = df.loc[df['datetime'] == entry_time, 'indexer'].iloc[0]
            st_index += 1
            while current_index < en_index:
                trade = {
                    'Stock Name': stock_name,
                    'Trade Type': 'LONG',
                    'Entry Date': df.iloc[current_index]['datetime'],
                    'Entry Price': df.iloc[current_index]['close'],
                    'Exit Date': df.iloc[st_index]['datetime'],
                    'Exit Price': df.iloc[st_index]['close'],
                    'Points Captured': df.iloc[st_index]['close'] - df.iloc[current_index]['close'],
                    'Qty': qty,
                    'PnL': (df.loc[st_index]['close'] - df.iloc[current_index]['close']) * qty,
                    'ROI%': ((df.loc[st_index]['close'] - df.iloc[current_index]['close']) * qty / portfolio_value) * 100,
                    'Pct Change on Stock': ((df.loc[st_index]['close'] - df.iloc[current_index]['close']) / df.loc[current_index]['close']) * 100,
                    'Trade Year': entry_time.year,
                    'Trade Number': f'{trade_number}L',
                }
                # print(trade)
                current_index += 1
                st_index += 1
    
                tradebook.append(trade)
            trade_number += 1
                
            entry_price = 0
            exit_price = 0
            points = 0
            remark = ''
            in_trade = False

    return pd.DataFrame(tradebook)
```

```python
def stats(stock_name, df):
    print(stock_name)
    stats = pd.DataFrame(index=range(2012, 2025), columns=['Total ROI', 'Total Trades', 'Win Rate', 'Avg Profit% per Trade', 'Avg Loss% per Trade', 'Max Drawdown', 'ROI/DD Ratio'])
    
    # Iterate over each year
    for year in range(2012, 2025):
        # Filter trades for the current year
        year_trades = df[(df['Trade Year'] == year)]
        
        # Calculate total ROI
        total_roi = year_trades['ROI%'].sum()
        
        # Calculate total number of trades
        total_trades = len(year_trades)
        
        # Calculate win rate
        win_rate = (year_trades['ROI%'] > 0).mean()*100
        
        # Calculate average profit per trade
        avg_profit = year_trades[year_trades['ROI%'] > 0]['ROI%'].mean()
        
        # Calculate average loss per trade
        avg_loss = year_trades[year_trades['ROI%'] < 0]['ROI%'].mean()
        
        # Calculate maximum drawdown
        max_drawdown = (year_trades['ROI%'].cumsum() - year_trades['ROI%'].cumsum().cummax()).min()
        
        # Calculate ROI/DD ratio
        roi_dd_ratio = total_roi / abs(max_drawdown)
        
        # Store the statistics in the DataFrame
        stats.loc[year] = [total_roi, total_trades, win_rate, avg_profit, avg_loss, max_drawdown, roi_dd_ratio]
    
    # Calculate overall statistics
    overall_total_roi = stats['Total ROI'].sum()
    overall_total_trades = stats['Total Trades'].sum()
    overall_win_rate = (df['ROI%'] > 0).mean() * 100
    overall_avg_profit = df[df['ROI%'] > 0]['ROI%'].mean()
    overall_avg_loss = df[df['ROI%'] < 0]['ROI%'].mean()
    overall_max_drawdown = (df['ROI%'].cumsum() - df['ROI%'].cumsum().cummax()).min()
    overall_roi_dd_ratio = overall_total_roi / abs(overall_max_drawdown)
    
    # Store the overall statistics in the DataFrame
    stats.loc['Overall'] = [overall_total_roi, overall_total_trades, overall_win_rate, overall_avg_profit, overall_avg_loss, overall_max_drawdown, overall_roi_dd_ratio]
    return stats

results = stats(current_stock, tb_pandas_per_month)
results
```
