```python
import pandas as pd
import polars as pl
pd.set_option("display.max_rows", 25_000)
pd.set_option("display.max_columns", 500)
pl.Config.set_tbl_cols(500)
pl.Config.set_tbl_rows(10_000)

pd.options.display.float_format = "{:.4f}".format
```

```python
df1 = pd.read_csv('MTrend Opt Buying 4 indices.csv')
df1['MTM'] = df1['Qty'] * df1['Points Captured']
df1.head(10)
```

```python
df1x = df1.groupby('Expiry', as_index=False)['MTM'].sum()
df1x.head(10)
```

```python
df1x.rename(columns={'Expiry': 'Date'}, inplace=True)
# df1 = df1[['Date', 'PnL']]
df1x['Date'] = pd.to_datetime(df1x['Date'], format='mixed', dayfirst=True)
df1x.head(10)
```

```python
df2 = pd.read_csv('MTrend Day-Wise MTM Sheet.csv')
df2.drop(columns=['MTM'], inplace=True)
df2.rename(columns={'MTM1': 'MTM', 'DATETIME': 'Date'}, inplace=True)
df2['Date'] = pd.to_datetime(df2['Date'], format='mixed', dayfirst=True)
df2.head(10)
```

```python
merged_df = pd.merge(df1x, df2, on='Date', how='outer', suffixes=('_opt_buy', '_pos')).fillna(0)
merged_df['Date'] = pd.to_datetime(merged_df['Date'])

merged_df_sorted = merged_df.sort_values(by='Date', ascending=True)
merged_df
```

```python
correlation = merged_df['MTM_opt_buy'].corr(merged_df['MTM_pos'])
correlation
```
